```python
from wykop import WykopAPI
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import time
from collections import deque
import keyboard
from pyvis.network import Network
import igraph as ig
import leidenalg as le

matplotlib.use('TkAgg')
```

```python
# ładowanie grafu
g = nx.read_gml("graph.gml")
g_ig = ig.Graph.from_networkx(g)
clusters = le.find_partition(g_ig, le.CPMVertexPartition, resolution_parameter=0.1)
print(f"Społeczności {clusters.membership}")
print(f"Liczba społeczności {len(clusters.membership)}")
```

```python
for node, community in zip(g.nodes(), clusters.membership):
    g.nodes[node]['community'] = community

nx.write_gml(g, "graph_with_communities.gml")

colors = [clusters.membership[i] for i in range(len(g))]
pos = nx.spring_layout(g)
nx.draw(g, pos, node_color=colors, with_labels=True, cmap=plt.cm.tab20, arrows=True)
plt.show
```
